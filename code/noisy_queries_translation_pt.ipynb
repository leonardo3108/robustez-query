# Obtenção dos textos das queries originais

```python
! wget -nc https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/queries-originals.txt
```

```python
query = {}
ids = []

for line in open('queries-originals.txt'):
    fields = line.strip().split()
    id = fields[0]
    ids.append(id)
    query[id] = {}
    query[id]['original'] = ' '.join(fields[1:])
```

```python
query
```

```python
ids[:10]
```

# Obtenção dos textos das queries traduzidas para português do Brasil

Queries traduzidas manualmente pelo Google Tradutor: https://translate.google.com.br/?hl=pt-BR&tab=TT&sl=en&tl=pt&op=translate

```python
! wget -nc https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/queries-tradutor-pt.txt
```

```python
for line in open('queries-tradutor-pt.txt'):
    fields = line.strip().split()
    id = fields[0]
    query[id]['translated-pt-br'] = ' '.join(fields[1:])
```

```python
query
```

# Obtenção dos textos das queries ruidosas (traduzidas para português do Brasil e depois de volta para  o inglês)

Queries traduzidas para português manualmente pelo Google Tradutor: https://translate.google.com.br/?hl=pt-BR&tab=TT&sl=en&tl=pt&op=translate

Depois traduzidas de volta para o inglês manualmente também pelo Google Tradutor: https://translate.google.com.br/?hl=pt-BR&tab=TT

```python
! wget -nc https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/noisy-queries-translation-pt.txt
```

```python
for line in open('noisy-queries-translation-pt.txt'):
    fields = line.strip().split()
    id = fields[0]
    query[id]['noisy-translation-pt'] = ' '.join(fields[1:])
```

```python
query
```

# Geração de DataFrame e exportação de CSV

```python
import pandas as pd

df = pd.DataFrame(query).T
df
```

```python
import csv

df.to_csv('noisy-queries-translation-pt.csv', header=True, index=True, index_label='Query#', quoting = csv.QUOTE_NONNUMERIC)
```

# Composição do arquivo mestre de queries ruidosas

```python
import requests

req = requests.get('https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/noisy-queries.txt')
if req.status_code != 200:
    df[['original', 'noisy-translation-pt']].to_csv('noisy-queries.csv', header=True, index=True, index_label='Query#', quoting = csv.QUOTE_NONNUMERIC)
```
