# Discounted Cumulative Gain (DCG@10)

## Obtenção da IDCG@10

```python
! wget -nc https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/idcg.csv
```

```python
import pandas as pd

df_DCG = pd.read_csv('idcg.csv')
df_DCG.rename(columns={'query': 'Query#', ' idcg': 'IDCG@10'}, inplace=True)
df_DCG
```

## Obtenção do DCG@10 do BM25

```python
! wget -nc https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/dcg_bm25_original.csv
```

```python
df = pd.read_csv('dcg_bm25_original.csv')
df
```

```python
df_DCG = df_DCG.join(df, rsuffix='_BM25')
del df_DCG['query']
df_DCG.rename(columns={' dcg@10': 'BM25_DCG@10'}, inplace=True)
df_DCG
```

## Obtenção do DCG@10 do BM25 + Rerank (MonoT5)

```python
! wget -nc https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/dcg10_rerank_monoT5_original.csv
```

```python
df = pd.read_csv('dcg10_rerank_monoT5_original.csv')
df
```

```python
df_DCG = df_DCG.join(df)
del df_DCG['query']
df_DCG.rename(columns={' dcg@10': 'Rerank_DCG@10'}, inplace=True)
df_DCG
```

## Obtenção do DCG@10 da busca densa (documentos com julgamentos)


> haystack.retriever.dense.DensePassageRetriever:

> query_embedding_model='facebook/dpr-question_encoder-single-nq-base'

> passage_embedding_model='facebook/dpr-ctx_encoder-single-nq-base'

```python
! wget -nc https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/dcg10_dense_dpr_mmpr_com_julgamento.csv
```

```python
df = pd.read_csv('dcg10_dense_dpr_mmpr_com_julgamento.csv')
df
```

```python
df_DCG = df_DCG.join(df)
del df_DCG['query']
df_DCG.rename(columns={' dcg@10': 'Dense_DCG@10'}, inplace=True)
df_DCG
```

## Obtenção dos textos das queries

```python
! wget -nc https://raw.githubusercontent.com/leonardo3108/robustez-query/main/data/queries-originals.txt
```

```python
queries = []

for query in open('queries-originals.txt'):
    fields = query.strip().split()
    queries.append(' '.join(fields[1:]))
```

```python
queries
```

```python
df_DCG['Text'] = queries
df_DCG
```

## Salvando os dados

```python
import csv

df_DCG.to_csv('dcg10_originals.csv', header=True, index=False, quoting = csv.QUOTE_NONNUMERIC)
```

```python
from google.colab import files

files.download("dcg10_originals.csv") 
```
