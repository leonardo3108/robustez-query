```python
judging = {}
for i, line in enumerate(open('2020qrels-pass.txt')):
    query, _, pid, eval = line.rstrip().split()
    judging[(query, pid)] = eval
```

```python
print('query pid judging')
for i, (query, pid) in enumerate(judging.keys()):
    print(query, pid, judging[(query, pid)])
    if i >= 100: break
```

```python
query_judgings = {}
for i, (query, pid) in enumerate(judging.keys()):
    if query not in query_judgings:
        query_judgings[query] = []
    query_judgings[query].append(int(judging[(query, pid)]))

print('Judgings per query:')
for query in query_judgings.keys():
    print('{}: {}'.format(query, query_judgings[query]))
```

```python
best10judgings = {}

print('Best 10 judgings per query:')
for query in query_judgings.keys():
    best10judgings[query] = sorted(query_judgings[query], reverse = True)[:10]
    print('{}: {}'.format(query, best10judgings[query]))
```

```python
import math

idcg = {}

print('IDCG calculation per query:')
print('query i j idcg acumulated')
for query in best10judgings.keys():
    acumulated_idcg = 0
    for i, judging in enumerate(best10judgings[query]):
        judging_idcg = (2**judging - 1) * math.log(2) / math.log(i+2)
        acumulated_idcg += judging_idcg
        print(query, i+1, judging, judging_idcg, acumulated_idcg)
    idcg[query] = acumulated_idcg            
```

```python
print('Query: IDCG')
for query in idcg.keys():
    print('{}: {}'.format(query, idcg[query]))
```

```python
w = open('idcg.csv', 'w')
for query in idcg.keys():
    w.write('{}, {}\n'.format(query, idcg[query]))
w.close()
```
